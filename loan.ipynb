```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df[' education'] = le.fit_transform(df[' education'])
```

```python
df.head()
```

```python
df[' self_employed']= pd.get_dummies(df[' self_employed'], drop_first=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df, columns =[' self_employed'])
```

```python
df.head()
```

```python
cat_cols = [' education', ' self_employed']
for col in cat_cols:
    cat_cols = [col for col in cat_cols if col in df.columns]
    for col in cat_cols:
        df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
dtc = DecisionTreeClassifier(max_depth=4, random_state=50)
x = df.drop(' loan_status', axis=1)
y = df[' loan_status']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=43)
dtc.fit(x_train, y_train)


```

```python
pred_y = dtc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
print("accuracy score: ", accuracy_score(y_test, pred_y))
```

```python
from sklearn.metrics import classification_report
print("classification report: \n", classification_report(y_test, pred_y))
```

```python
from sklearn import tree
plt.figure(figsize=(20,10))
tree.plot_tree(dtc, feature_names=x.columns, class_names=['approved', 'rejected'], filled=True)
plt.show()
```
